Analyzing Companies Finanical Health w/ Nasdaq Data

This project explores company finanical data from the Nasdaq Data Link MER/F1 dataset, which includes various metrics from balance sheets, income statements, and cash flow. It aims to evaluate the financial performance of companies over a period of time and compare the AET (accurred expense turnover) through the retrieval and cleaning of JSON data from the API, the subsequent data analysis w/ Pandas and data visualization w/ Matplotlib

```python
# Processing the data into a dataframe
URL = "https://data.nasdaq.com/api/v3/datatables/MER/F1.json"
api_key = config.API_KEY

parameters = {
    'api_key': api_key,
    'qopts.per_page': 10000
}
response = requests.get(URL, params = parameters)
json_data = response.json()

data = json_data["datatable"]["data"]

columns = json_data["datatable"]["columns"]

columns_name = [col["name"]for col in columns]

df_metric = pd.DataFrame(data, columns=columns_name)

df_metric
```

```python
necessary_columns=['reportid','reportdate','reporttype','amount','longname','country','region','indicator','statement']


#It is essential to identify companies, but if we have one indicator, 
# we do not need multiple, such as compnumber. It is necessary to 
# follow the money, the time and place. Thus, these var are vital. 
```

```python
df_metric = df_metric[necessary_columns].copy()

#Extracting the financial data for accrued expenses turnover
df_filtered = df_metric.loc[df_metric["indicator"] == "Accrued Expenses Turnover"].copy()

df_filtered.describe()
```

```python
def country_name_converter(name):
    country_mapping = {
    'USA': 'United States of America',
    'DEU': 'Germany',
    'JPN': 'Japan',
    'CYM': 'Cayman Islands',
    'BHS': 'Bahamas',
    'IRL': 'Ireland'
   
}
    return country_mapping.get(name, "Unknown")
#Changing the countries to their full names
df_filtered.loc[:,"country_name"] = df_filtered["country"].apply(country_name_converter)
updated_df = df_filtered.rename(columns={"longname": "company_name",
                                         "reportid" : "report_id",
                                         "reportdate" : "report_date",
                                         "reporttype" : "report_type",
                                        }, inplace=False)

updated_df["country_name"].value_counts()
```

```python
import matplotlib.pyplot as plt

#Filter for 2010-2015
updated_df["report_date"] = pd.to_datetime(updated_df["report_date"])
updated_df = updated_df.loc[(updated_df["report_date"].dt.year >= 2010) & (updated_df["report_date"].dt.year <= 2015)]

#Using relevant columns to plot financial trends
relevant_data = updated_df[["company_name", "report_date", "amount"]].copy()

plt.figure(figsize=(12,6))
for unique_company in relevant_data["company_name"].unique():
    company_data = relevant_data.loc[unique_company == relevant_data["company_name"]]
    plt.plot(company_data["report_date"], company_data["amount"], label = unique_company)
    plt.title("Analysis of the Accrued Expenses Turnover (2010-2015)")
    plt.xlabel("Report Dates")
    plt.ylabel("Amount")
    plt.legend(
        title=unique_company,              
        bbox_to_anchor=(1.05, 1),     
        loc="upper left" 
        
    )
plt.show()
```

```python
#Group data by region 
country_avg = updated_df.groupby("country_name")["amount"].mean()
country_avg.plot(kind = "bar", figsize = (12,7))

#Graph figurations 
plt.title("Financial Perfomance by Countries")
plt.xlabel("Countries")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
